```python
# %matplotlib notebook
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merged_df = pd.DataFrame(pd.merge(mouse_metadata, study_results, on='Mouse ID', how='outer'))
# Display the data table for preview
merged_df
```

```python
# Checking the number of mice.
print(merged_df['Mouse ID'].nunique())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
duplicates = merged_df[merged_df.duplicated(['Mouse ID','Timepoint'],keep=False)]
dmid = duplicates['Mouse ID'].unique()
dmid
```

```python
# reset index to mouse id
merged_df = merged_df.set_index('Mouse ID')
merged_df
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
dmid_data = merged_df.loc[dmid]
dmid_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = merged_df
clean_df.drop(clean_df.loc[dmid].index,inplace=True)
clean_df = clean_df.reset_index()
clean_df
```

```python
# Checking the number of mice in the clean DataFrame.
print(clean_df['Mouse ID'].nunique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance,
# standard deviation, and SEM of the tumor volume for each regimen
regimen_data = clean_df.groupby(['Drug Regimen'])
reg_groups_mean = regimen_data['Tumor Volume (mm3)'].mean()
reg_groups_std = regimen_data['Tumor Volume (mm3)'].std()
reg_groups_median = regimen_data['Tumor Volume (mm3)'].median()
reg_groups_var = regimen_data['Tumor Volume (mm3)'].var()
reg_groups_sem = regimen_data['Tumor Volume (mm3)'].sem()
# This method is the most straighforward,
# creating multiple series and putting them all together at the end.
summary_sts = pd.DataFrame({'Mean': reg_groups_mean, 'Standard Deviation':reg_groups_std,
                            'Median': reg_groups_median, 'Variance': reg_groups_var, 'SEM':reg_groups_sem})
summary_sts
```
